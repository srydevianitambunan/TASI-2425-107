## Evaluasi Long-tail

```python
def LT_Ratio(interactions, alpha=0.8):
    """
    Fungsi untuk menghitung tingkat long-tailedness dari suatu dataset.
    :param interactions: Data jumlah interaksi per item
    :param alpha: Proporsi yang digunakan untuk menghitung LT Ratio (default: 0.8)
    """
    num_items = len(interactions)

    sorted_interactions = np.sort(interactions)[::-1]  # Urutkan interaksi dari besar ke kecil

    sum_1 = 0
    Q_80 = 0  # Jumlah kategori yang berkontribusi pada 80% interaksi

    # Menghitung jumlah kategori yang berkontribusi pada 80% interaksi
    for i in range(num_items):
        sum_1 += sorted_interactions[i]
        if sum_1 >= alpha * interactions.sum():  # Jika sudah mencapai 80% dari total interaksi
            Q_80 = i + 1  # Menyimpan jumlah kategori yang berkontribusi pada 80%
            break

    # C adalah total jumlah kategori
    C = num_items

    # Menghitung Pareto-LT Ratio sesuai dengan rumus C - Q(0.8) / Q(0.8)
    LT_ratio = (C - Q_80) / Q_80
    return LT_ratio

def Imbalance_factor(labels):
    """
    Fungsi untuk menghitung faktor ketidakseimbangan dalam dataset.
    :param labels: Data jumlah interaksi per item
    """
    # Hitung jumlah interaksi untuk setiap kategori
    num_labels_each_class = np.array([(labels == i).sum().item() for i in range(labels.max() + 1)])
    # Hapus kategori dengan jumlah nol
    num_labels_each_class = np.array([i for i in num_labels_each_class if i > 0])

    # Urutkan jumlah interaksi dari besar ke kecil
    sorted_num_labels_each_class = np.sort(num_labels_each_class)[::-1]

    # Imbalance Factor: Rasio jumlah interaksi tertinggi dan terendah
    IF = sorted_num_labels_each_class[0] / sorted_num_labels_each_class[-1]

    # Cetak hasil
    print("Jumlah kategori (non-zero):", len(num_labels_each_class))
    print("Imbalance Factor: %.3f" % IF)

def Gini(labels):
    """
    Fungsi untuk menghitung koefisien Gini dari distribusi label.
    :param labels: Data jumlah interaksi per item
    """
    labels = np.array(labels, dtype=np.int64)
    total = 0
    for i, xi in enumerate(labels[:-1], 1):
        total += np.sum(np.abs(xi - labels[i:]))
    gini = total / (len(labels) ** 2 * np.mean(labels))
    return gini
```

```python
def calculate_sparsity(df, user_col='reviewerID', item_col='asin'):
    """
    Menghitung tingkat sparsity dari dataset.
    :param df: DataFrame dengan interaksi pengguna-item
    :param user_col: Nama kolom pengguna (default: 'reviewerID')
    :param item_col: Nama kolom item (default: 'asin')
    :return: Persentase sparsity
    """
    num_users = df[user_col].nunique()
    num_items = df[item_col].nunique()
    num_interactions = len(df)

    sparsity = 1 - (num_interactions / (num_users * num_items))
    return sparsity

#  Load dataset
df_movies = pd.read_csv('Attributes_Movies_and_TV.csv')
df_books = pd.read_csv('Attributes_Books.csv')

# Menghitung sparsity keseluruhan
sparsity_movies = calculate_sparsity(df_movies)
sparsity_books = calculate_sparsity(df_books)
```

```python
# Menentukan kategori head dan tail berdasarkan jumlah interaksi
def categorize_items(df, item_col='asin', threshold=0.2):
    """
    Mengkategorikan item menjadi head dan tail berdasarkan distribusi interaksi.
    :param df: DataFrame dengan interaksi pengguna-item
    :param item_col: Nama kolom item (default: 'asin')
    :param threshold: Proporsi head dalam distribusi (default: 0.2)
    :return: DataFrame kategori head dan tail
    """
    item_counts = df[item_col].value_counts()
    total_interactions = item_counts.sum()
    
    # Tentukan batas jumlah interaksi untuk head
    head_cutoff = item_counts.cumsum() / total_interactions <= threshold
    head_items = item_counts[head_cutoff].index
    tail_items = item_counts[~head_cutoff].index

    return df[df[item_col].isin(head_items)], df[df[item_col].isin(tail_items)]

# Kategorisasi head dan tail untuk Movies
df_movies_head, df_movies_tail = categorize_items(df_movies, item_col='asin', threshold=0.2)

# Kategorisasi head dan tail untuk Books
df_books_head, df_books_tail = categorize_items(df_books, item_col='asin', threshold=0.2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi untuk menampilkan distribusi Pareto
def plot_pareto(df, dataset_name, alpha=0.8):
    """
    Visualisasi distribusi Pareto (Long-Tailedness Ratio).

    Args:
        df (DataFrame): Dataset yang berisi jumlah interaksi per item.
        dataset_name (str): Nama dataset.
        alpha (float): Proporsi data untuk head (default: 0.8 untuk 80%).
    """
    jumlah_interaksi_per_produk = df.groupby('asin').size().reset_index(name='jumlah_interaksi')
    labels = jumlah_interaksi_per_produk['jumlah_interaksi'].values

    sorted_labels = np.sort(labels)[::-1]
    total_interactions = sorted_labels.sum()
    cumulative_interactions = np.cumsum(sorted_labels) / total_interactions
    head_cutoff = np.where(cumulative_interactions >= alpha)[0][0]

    plt.figure(figsize=(10, 6))
    plt.plot(range(len(sorted_labels)), cumulative_interactions, label='Cumulative Distribution')
    plt.axvline(x=head_cutoff, color='red', linestyle='--', label=f'Pareto Head ({alpha * 100:.0f}%)')
    plt.xlabel('Rank (Items Sorted by Interactions)')
    plt.ylabel('Cumulative Percentage of Interactions')
    plt.title(f'Pareto Long-Tailedness Distribution - {dataset_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualisasi Pareto Ratio untuk Amazon Movies
plot_pareto(df_movies, "Amazon Movies", alpha=0.8)
```

```python
def plot_imbalance_with_details(df, dataset_name):
    """
    Visualisasi distribusi ketimpangan (Imbalance Factor) dengan penjelasan tambahan.

    Args:
        df (DataFrame): Dataset yang berisi jumlah interaksi per item.
        dataset_name (str): Nama dataset.
    """
    jumlah_interaksi_per_produk = df.groupby('asin').size().reset_index(name='jumlah_interaksi')
    labels = jumlah_interaksi_per_produk['jumlah_interaksi'].values
    sorted_labels = np.sort(labels)[::-1]

    max_interaction = sorted_labels[0]
    min_interaction = sorted_labels[-1]

    plt.figure(figsize=(12, 8))
    plt.bar(range(len(sorted_labels)), sorted_labels, color='skyblue', edgecolor='blue')

    plt.bar(0, max_interaction, color='orange', label=f'Head (Max: {max_interaction} interactions)')
    plt.bar(len(sorted_labels)-1, min_interaction, color='red', label=f'Tail (Min: {min_interaction} interactions)')

    plt.yscale('log')
    plt.xlabel('Rank (Products Sorted by Interactions)', fontsize=12)
    plt.ylabel('Number of Interactions (Log Scale)', fontsize=12)
    plt.title(f'Imbalance Factor Distribution - {dataset_name} (Log Scale)', fontsize=14)

    plt.annotate(f'Max: {max_interaction}',
                 xy=(0, max_interaction),
                 xytext=(50, max_interaction * 1.1),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 fontsize=10,
                 color='black')

    plt.annotate(f'Min: {min_interaction}',
                 xy=(len(sorted_labels)-1, min_interaction),
                 xytext=(len(sorted_labels)-200, min_interaction * 10),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 fontsize=10,
                 color='black')

    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Visualisasi Imbalance Factor untuk Amazon Movies
plot_imbalance_with_details(df_movies, "Amazon Movies")
```

```python
def plot_gini(df, dataset_name):
    """
    Visualisasi distribusi untuk Gini Coefficient.

    Args:
        df (DataFrame): Dataset yang berisi jumlah interaksi per item.
        dataset_name (str): Nama dataset.
    """
    jumlah_interaksi_per_produk = df.groupby('asin').size().reset_index(name='jumlah_interaksi')
    labels = jumlah_interaksi_per_produk['jumlah_interaksi'].values

    sorted_labels = np.sort(labels)
    cumulative_proportions = np.cumsum(sorted_labels) / np.sum(sorted_labels)

    x = np.linspace(0, 1, len(cumulative_proportions))
    y = cumulative_proportions

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Lorenz Curve', color='blue')
    plt.plot([0, 1], [0, 1], linestyle='--', color='red', label='Equality Line (Gini = 0)')
    plt.fill_between(x, y, x, color='skyblue', alpha=0.5)
    plt.xlabel('Cumulative Share of Products')
    plt.ylabel('Cumulative Share of Interactions')
    plt.title(f'Gini Coefficient Visualization - {dataset_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualisasi Gini Coefficient untuk Amazon Movies
plot_gini(df_movies, "Amazon Movies")
```

```python
# Fungsi untuk menampilkan distribusi Pareto
def plot_pareto(df, dataset_name, alpha=0.8):
    """
    Visualisasi distribusi Pareto (Long-Tailedness Ratio).

    Args:
        df (DataFrame): Dataset yang berisi jumlah interaksi per item.
        dataset_name (str): Nama dataset.
        alpha (float): Proporsi data untuk head (default: 0.8 untuk 80%).
    """
    jumlah_interaksi_per_produk = df.groupby('asin').size().reset_index(name='jumlah_interaksi')
    labels = jumlah_interaksi_per_produk['jumlah_interaksi'].values

    sorted_labels = np.sort(labels)[::-1]
    total_interactions = sorted_labels.sum()
    cumulative_interactions = np.cumsum(sorted_labels) / total_interactions
    head_cutoff = np.where(cumulative_interactions >= alpha)[0][0]

    plt.figure(figsize=(10, 6))
    plt.plot(range(len(sorted_labels)), cumulative_interactions, label='Cumulative Distribution')
    plt.axvline(x=head_cutoff, color='red', linestyle='--', label=f'Pareto Head ({alpha * 100:.0f}%)')
    plt.xlabel('Rank (Items Sorted by Interactions)')
    plt.ylabel('Cumulative Percentage of Interactions')
    plt.title(f'Pareto Long-Tailedness Distribution - {dataset_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualisasi Pareto Ratio untuk Amazon Books
plot_pareto(df_books, "Amazon Books", alpha=0.8)
```

```python
def plot_gini(df, dataset_name):
    """
    Visualisasi distribusi untuk Gini Coefficient.

    Args:
        df (DataFrame): Dataset yang berisi jumlah interaksi per item.
        dataset_name (str): Nama dataset.
    """
    jumlah_interaksi_per_produk = df.groupby('asin').size().reset_index(name='jumlah_interaksi')
    labels = jumlah_interaksi_per_produk['jumlah_interaksi'].values

    sorted_labels = np.sort(labels)
    cumulative_proportions = np.cumsum(sorted_labels) / np.sum(sorted_labels)

    x = np.linspace(0, 1, len(cumulative_proportions))
    y = cumulative_proportions

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Lorenz Curve', color='blue')
    plt.plot([0, 1], [0, 1], linestyle='--', color='red', label='Equality Line (Gini = 0)')
    plt.fill_between(x, y, x, color='skyblue', alpha=0.5)
    plt.xlabel('Cumulative Share of Products')
    plt.ylabel('Cumulative Share of Interactions')
    plt.title(f'Gini Coefficient Visualization - {dataset_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualisasi Gini Coefficient untuk Amazon Books
plot_gini(df_books, "Amazon Books")
```
