## Amazon-KG-BiTGCF-Book
Training Loss and Validation BiTGCF-Amazon-KG-Book per Epoch

```python
import re
import matplotlib.pyplot as plt


# Path ke file log
log_file = '/content/drive/MyDrive/Google Colab TA/Log-Model-TA/Amazon-KG/BiTGCF.log'

# Variabel untuk menyimpan data
epochs = []
train_loss1 = []
train_loss2 = []
valid_scores = []

# Baca file log
with open(log_file, "r") as f:
    for line in f:
        # Cari informasi epoch
        epoch_match = re.search(r"epoch (\d+) training", line)
        if epoch_match:
            epoch = int(epoch_match.group(1))
            epochs.append(epoch)

        # Cari train_loss1 dan train_loss2
        loss_match = re.search(r"train_loss1: ([\d.]+), train_loss2: ([\d.]+)", line)
        if loss_match:
            train_loss1.append(float(loss_match.group(1)))
            train_loss2.append(float(loss_match.group(2)))

        # Cari valid_score
        valid_match = re.search(r"valid_score: ([\d.]+)", line)
        if valid_match:
            valid_scores.append(float(valid_match.group(1)))

# Plot hasil training
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_loss1, label="Train Loss 1", marker="o", markersize=4, color="black", linestyle="-")
plt.plot(epochs, train_loss2, label="Train Loss 2", marker="s", markersize=4, color="gray", linestyle="--")
plt.plot(epochs, valid_scores, label="Validation Score", marker="d", markersize=4, color="dimgray", linestyle="-.")

# Pengaturan tampilan
plt.xlabel("Epoch")
plt.ylabel("Loss / Score")
plt.title("Training Loss and Validation CKE-Gemini-Books")
plt.legend()

# Cek apakah data ada
print(f"Epochs: {epochs}")
print(f"Train Loss1: {train_loss1}")
print(f"Train Loss2: {train_loss2}")
print(f"Valid Score: {valid_scores}")
```

## KG-BPR Book

```python
import matplotlib.pyplot as plt
import re

# Fungsi untuk membaca data dari file log
def extract_data_from_log(log_file):
    epochs = []
    train_loss = []
    valid_score = []

    with open(log_file, 'r') as file:
        epoch_num = 0
        for line in file:
            # Mencari baris yang mengandung 'train loss' dengan regex
            train_loss_match = re.search(r'epoch (\d+) training.*train loss: ([\d\.]+)', line)
            if train_loss_match:
                epochs.append(epoch_num)  # Tambahkan epoch
                train_loss.append(float(train_loss_match.group(2)))  # Ambil nilai train loss

            # Mencari baris yang mengandung 'valid_score' dengan regex
            valid_score_match = re.search(r'epoch (\d+) evaluating.*valid_score: ([\d\.]+)', line)
            if valid_score_match:
                valid_score.append(float(valid_score_match.group(2)))  # Ambil nilai valid score

            # Increment epoch setiap kali menemukan evaluasi untuk epoch berikutnya
            if 'evaluating' in line:
                epoch_num += 1

    return epochs, train_loss, valid_score

# Ambil data dari file log (ganti 'your_log_file.txt' dengan nama file log yang sebenarnya)
log_file = log_file = '/content/drive/MyDrive/Google Colab TA/Log-Model-TA/Amazon-KG/BPR Books Amazon KG.log'
epochs, train_loss, valid_score = extract_data_from_log(log_file)

# Cek apakah data ada
print(f"Epochs: {epochs}")
print(f"Train Loss: {train_loss}")
print(f"Valid Score: {valid_score}")

# Plot grafik jika data ada
if len(epochs) > 0:
    plt.figure(figsize=(10,6))

    # Plot train loss
    plt.plot(epochs, train_loss, label='Train Loss', color='black', marker='o')

    # Plot valid score
    plt.plot(epochs, valid_score, label='Valid Score', color='black', marker='x')

    # Menambahkan label dan judul
    plt.title('Training Loss and Validation BPR-Amazon-KG-Book')
    plt.xlabel('Epoch')
    plt.ylabel('Value')
    plt.legend()

    # Tidak menampilkan grid
    # plt.grid(True)  # Hapus atau komentar baris ini untuk menghilangkan grid

    # Menampilkan grafik
    plt.show()
else:
    print("Tidak ada data untuk diplot.")
```
